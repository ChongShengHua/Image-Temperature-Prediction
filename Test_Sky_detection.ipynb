# 1.0 Test Sky Detection function


```
import sky_detect
import numpy as np
import cv2
import pandas as pd
import os

df1 = pd.read_csv('selected_img.csv')
has_sky_img = pd.concat([df1[df1['CamId'] == 1093].head(200), df1[df1['CamId'] == 10066].head(200), df1[df1['CamId'] == 10870].head(200)]).reset_index()
has_sky_img = has_sky_img.Filename.to_list()
```

```
sky_detect_init = sky_detect.sky_detect()

folder = 'D:/y3s1/VISUAL INFORMATION PROCESSING/348315_Project/1093'
detected = 0
not_detected = 0
i = 0

result = pd.DataFrame(columns = ['Filename','sky_detected'])


for filename in os.listdir(folder):
    if filename in has_sky_img:
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            if sky_detect_init.detect_sky(img):
                result.loc[i] = [filename, 'yes']
                detected += 1

            else:
                result.loc[i] = [filename, 'no']
                not_detected += 1

            print(i)
            i += 1


result.to_csv('result_detected_has_sky_1093.csv', encoding='utf-8', index=False)

```
